```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML

from smbbh_nu import SMBBH_NU
from plot_process import Plot_Result
```

## 1. Parameters

### 1-0. Potential

```python
def potential_1(constant_c, tmp_array, ratio_mass, r):
    ratio_term = constant_c*(tmp_array*ratio_mass/r)
    first_term = 1/(r + r**3)
    second_term = np.arctan(r)/(r**2)
    V = ratio_term*(first_term - second_term)
    return V
```

### 1-1. Case_1

```python
black_hole_mass = [0.5, 0.5]  # m1, m2
radius = 0.5
eccentricity = 0
constant_c = 0 
angles = [np.pi/6, np.pi/4, np.pi/6]  # omega, I, Omega
```

```python
experiment_1 = SMBBH_NU(black_hole_mass,
                        constant_c=constant_c,
                        radius=radius,
                        eccentricity=eccentricity,
                        angles=angles,
                        potential_function=potential_1)
result_1 = experiment_1.run()
```

### 1-2. Case_2

```python
black_hole_mass = [0.5, 0.5]  # m1(0.5), m2(0.5)
radius = 0.4  # 0.5
eccentricity = 0
constant_c = 0 
angles = [np.pi/6, np.pi/4, np.pi/6]  # omega, I, Omega
```

```python
experiment_2 = SMBBH_NU(black_hole_mass,
                        constant_c=constant_c,
                        radius=radius,
                        eccentricity=eccentricity,
                        angles=angles,
                        potential_function=potential_1)
result_2 = experiment_2.run()
```

### 1-3. Case_3

```python
black_hole_mass = [0.5, 0.5]  # m1(0.5), m2(0.5)
radius = 0.3  # 0.5
eccentricity = 0
constant_c = 0 
angles = [np.pi/6, np.pi/4, np.pi/6]  # omega, I, Omega
```

```python
experiment_3 = SMBBH_NU(black_hole_mass,
                        # time_length=8*10**3, 
                        constant_c=constant_c,
                        radius=radius,
                        eccentricity=eccentricity,
                        angles=angles,
                        potential_function=potential_1)
result_3 = experiment_3.run()
```

## Plot $E_{total}/E_{0}$

```python
case1_Et = result_1['total_energy']
case2_Et = result_2['total_energy']
case3_Et = result_3['total_energy']

case1_E0 = result_1['initial_energy']
case2_E0 = result_2['initial_energy']
case3_E0 = result_3['initial_energy']

time_length1 = result_1['time_length']
dt_len1 = np.linspace(0, time_length1, time_length1 - 1)
time_length2 = result_2['time_length']
dt_len2 = np.linspace(0, time_length2, time_length2 - 1)
time_length3 = result_3['time_length']
dt_len3 = np.linspace(0, time_length3, time_length3 - 1)


Et_E0_1 = case1_Et/case1_E0 
Et_E0_2 = case2_Et/case2_E0 
Et_E0_3 = case3_Et/case3_E0 
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(10, 4))
plt.subplot()
plt.plot(dt_len1, Et_E0_1, "-", color="red", label="case 1")
plt.plot(dt_len2, Et_E0_2, "-", color="green", label="case 2")
plt.plot(dt_len3, Et_E0_3, "-", color="darkblue", label="case 3")
plt.title(r'$E_{total}/E_{0}$', fontsize=20)
plt.xlabel("Time ($4.3*10^{15}$ years)", fontsize=10)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
