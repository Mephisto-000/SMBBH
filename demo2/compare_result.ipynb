# Compare $c = 0$、$c = 1\cdot 10^{3}$、$c = 1\cdot 10^{6}$、$c = c\cdot 10^{9}$

```python
%matplotlib widget
import os
import numpy as np
import matplotlib.pylab as plt

from smbbh_nu import SMBBH_NU
from plot_process import Plot_Result

plt.rcParams.update({'figure.max_open_warning': 0})
plt.rcParams["figure.figsize"] = [1.5 * i for i in plt.rcParams["figure.figsize"]]  # x2.5
```

```python
total_smbbh_mass = 1.2*(10**12)
black_hole_mass = [total_smbbh_mass/2, total_smbbh_mass/2]  # m1, m2
mu = 1 * (total_smbbh_mass/2 + total_smbbh_mass/2)
radius = 0.05
time_length = 1000  #1000
time_step = 0.0005  #0.0005
e = 0.5
rot_angles = [np.pi/6, np.pi/4, np.pi/6] # omega, I, Omega
```

```python
def potential_1(constant_c, tmp_array, ratio_mass, r):
    ratio_term = constant_c*(tmp_array*ratio_mass/r)
    first_term = 1/(r + r**3)
    second_term = np.arctan(r)/(r**2)
    V = ratio_term*(first_term - second_term)
    return V
```

## 1. $c = 0$

```python
constant_c1 = 0

experiment_1 = SMBBH_NU(black_holes_mass=black_hole_mass, 
                        constant_c=constant_c1, 
                        radius=radius, 
                        eccentricity=e, 
                        angles=rot_angles, 
                        potential_function=potential_1)
result_1 = experiment_1.all_result_output()
```

```python
all_data1 = result_1["rot_data"]

r1x1, r1y1, r1z1 = all_data1[0:, 0], all_data1[0:, 1], all_data1[0:, 2]
r2x1, r2y1, r2z1 = all_data1[0:, 3], all_data1[0:, 4], all_data1[0:, 5]
v1x1, v1y1, v1z1 = all_data1[0:, 6], all_data1[0:, 7], all_data1[0:, 8]
v2x1, v2y1, v2z1 = all_data1[0:, 9], all_data1[0:, 10], all_data1[0:, 11]
```

```python
r1_1 = np.sqrt(r1x1**2 + r1y1**2 + r1z1**2)
E1_1 = 0.5*(v1x1**2 + v1y1**2 + v1z1**2) - (constant_c1*(np.arctan(r1_1)/r1_1) + mu/r1_1)
r2_1 = np.sqrt(r2x1**2 + r2y1**2 + r2z1**2)
E2_1 = 0.5*(v2x1**2 + v2y1**2 + v2z1**2) - (constant_c1*(np.arctan(r2_1)/r2_1) + mu/r2_1)

E1_total = E1_1 + E2_1
```

## Plot $\Delta E = E_{t}/E_{0}$

```python
Et1_divi_E0 = E1_total/E1_total[0]
Et2_divi_E0 = E2_total/E2_total[0]
Et3_divi_E0 = E3_total/E3_total[0]
Et4_divi_E0 = E4_total/E4_total[0]
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(10, 5))
total_time = int(time_length / time_step)
dt_len = np.linspace(0, total_time, time_length)
plt.subplot()

plt.plot(dt_len, Et1_divi_E0, "--", color="darkblue", label=r"$c = 0$")
plt.plot(dt_len, Et2_divi_E0, "--", color="darkviolet", label=r"$c = 1\cdot 10^{3}$")
plt.plot(dt_len, Et3_divi_E0, "--", color="green", label=r"$c = 1\cdot 10^{6}$")
plt.plot(dt_len, Et4_divi_E0, "--", color="red", label=r"$c = 1\cdot 10^{9}$")

plt.xlabel("Time ($4.3*10^{15}$ years)", fontsize=10)
plt.ylabel(r"$\Delta E$", fontsize=10)
plt.title(r"$\Delta E = E_{t} / E_{0}$", fontsize=20)
plt.legend(loc="upper right", fontsize=20)

plt.show()
```
